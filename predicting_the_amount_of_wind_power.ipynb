```python
train[train['target'] < low]
```

### 6. 이상치로 판단된 target의 상한값 이상 데이터

상한값보다 큰 이상치는 꽤나 많이 나온다.

```python
train[train['target'] > high]
```

### 7. 이상치 유무에 따른 target과 feature들의 평균 비교

여러개의 그래프를 동시에 그리기 위해서 subplots()함수를 이용해 보자.

```python
plt.rc('font', family='Malgun Gothic')
fig=plt.figure(figsize=(20,10), dpi = 100)
axs=fig.subplots(2,4)

lm_features = ['temperature', 'pressure', 'humidity', 'wind_speed',
       'wind_direction', 'precipitation', 'snowing', 'cloudiness']

for i,j in enumerate(lm_features):

    x = ['target>high','target<=high']
    y = [train[train['target']>high][j].mean(), train[train['target']<=high][j].mean()]

    row = int(i/4)
    col = i%4


    axs[row][col].set_title("target X " + lm_features[i], fontsize= 20)
    axs[row][col].set_ylabel(lm_features[i]+"의 평균", fontsize= 15)
    axs[row][col].set_xticks([0,1])
    axs[row][col].set_xticklabels(['target>high','target<=high'], fontsize = 17)
    axs[row][col].bar(x, y)
    
    fig.tight_layout()

plt.show()
```

위의 그래프를 통해 풍속, 풍향, 강수량 등 일부 변수들은 이상치가 아닌 데이터에 비해 높은 값을 보이고 있다. 또한 눈이 내린 날은 한 번도 없음을 알 수 있다.

## 스테이지 7

### 1. pandas를 이용해 csv 파일 읽어오기


```python
import pandas as pd

train = pd.read_csv('data/wind_power/train.csv')
test = pd.read_csv('data/wind_power/test.csv')
submission = pd.read_csv('data/wind_power/sample_submission.csv')
```

### 2. 데이터 전처리


```python
# 카테고리형 변수 전처리
# 현재 동작을 유지하는 방법
train['snowing'] = train['snowing'].replace(True, 1).infer_objects(copy=False)
train['snowing'] = train['snowing'].replace(False, 0).infer_objects(copy=False)

test['snowing'] = test['snowing'].replace(True, 1).infer_objects(copy=False)
test['snowing'] = test['snowing'].replace(False, 0).infer_objects(copy=False)


# 정규화
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
MinMax_columns = ['temperature', 'pressure', 'humidity', 'wind_speed', 'wind_direction', 'precipitation', 'cloudiness']
train[MinMax_columns] = scaler.fit_transform(train[MinMax_columns])
test[MinMax_columns] = scaler.transform(test[MinMax_columns])

#이상치 제거
Q3 = train.describe()['target']['75%']
Q1 = train.describe()['target']['25%']
IQR = Q3 - Q1
high = Q3 + 1.5 * IQR

train_normal = train[train['target'] <= high]
train_normal
```

### 3. 독립/종속 변수 설정, train/valid 나누기


```python
x = train_normal.drop(['id', 'target'], axis=1)
y = train_normal['target']

from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

### 4. MAE 평가지표 정의하기


```python
from sklearn.metrics import mean_absolute_error

def evaluation(y_valid, pred):
    mae = mean_absolute_error(y_valid,pred) 
    return mae
```

### 5. GridSearchCV


GridSearchCV는 최적의 하이퍼파라미터 조합을 찾아준다.

```python
from sklearn.model_selection import GridSearchCV
```

### 6. RandomForest 고도화


* cv: 교차 검증 폴드 개수
* scoring: 모델 성능 평가 지표
* verbose: 모델 학습 시 진행 상황을 출력 -> 2: 간략한 정보, 1: 자세한 정보, 0: 출력 X

```python
from sklearn.ensemble import RandomForestRegressor

params = {'n_estimators' : [100, 150],'max_depth':[5,10]}

RF_model = GridSearchCV(RandomForestRegressor(),
                         param_grid=params, 
                         cv=3, 
                         scoring = 'neg_mean_absolute_error',
                         verbose=2)

RF_model.fit(x_train,y_train)
RF_pred = RF_model.predict(x_valid)
```

### 7. valid 셋 검증 결과


```python
scores_df = pd.DataFrame(RF_model.cv_results_)
display(scores_df[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score']])

print("최적의 점수 :",RF_model.best_score_)
print("최적의 하이퍼 파라미터 :",RF_model.best_params_)
```

### 8. valid셋 예측/평가


```python
estimator =RF_model.best_estimator_

pred = estimator.predict(x_valid) 
print('테스트 데이터 세트 MAE:{0:.4f}'.format(evaluation(y_valid,pred)))
```

### 9. 전체 train 데이터를 사용한 RandomForestRegressor 학습/예측


```python
from sklearn.ensemble import RandomForestRegressor

params = {'n_estimators' : [100, 150],'max_depth':[5,10]}

RF_model = GridSearchCV(RandomForestRegressor(random_state = 42),
                         param_grid=params, 
                         cv=3, 
                         scoring = 'neg_mean_absolute_error',
                         verbose=2)

RF_model.fit(x,y)
estimator =RF_model.best_estimator_
RF_pred_real = estimator.predict(test[test.columns[1:]]) 
```

### 10. test 셋 검증 결과


```python
print("최적의 점수 :",RF_model.best_score_)
print("최적의 하이퍼 파라미터 :",RF_model.best_params_)
```

### 11. submission 파일에 target 값 채우기 및 CSV 파일 저장

```python
submission['target'] = RF_pred_real
submission.head()
```
