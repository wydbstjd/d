# 쇼핑몰 지점별 매출액 예측 해커톤

* id : 샘플 아이디
* Store : 쇼핑몰 지점
* Date : 주 단위(Weekly) 날짜
* Temperature : 해당 쇼핑몰 주변 기온
* Fuel_Price : 해당 쇼핑몰 주변 연료 가격
* Promotion 1~5 : 해당 쇼핑몰의 비식별화된 프로모션 정보
* Unemployment : 해당 쇼핑몰 지역의 실업률
* IsHoliday : 해당 기간의 공휴일 포함 여부
* Weekly_Sales : 주간 매출액 (목표 예측값)

## VotingRegressor인 경우

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
train = pd.read_csv('data/shopping/train.csv')
test = pd.read_csv('data/shopping/test.csv')
submission = pd.read_csv('data/shopping/sample_submission.csv')
```

train 데이터셋에 대한 구조를 살펴보자.

총 6255개의 행과 13개의 열로 구성되어 있고 5개의 Promotion 열에는 결측값이 존재한다.

```python
display(train.info())
display(train.describe())
train.head()
```

1. Date 피처를 각각 year, month, Week로 나눠보자.
2. IsHoliday bool 타입을 True면 1, False면 0을 바꿔보자.
3. Promotion 열의 결측치를 0으로 채워준 뒤, 모두 합한 피처를 만들자.
4. 이상치 탐지 및 제거를 해보자.

```python
train['Date'] = pd.to_datetime(train['Date'], format="%d/%m/%Y")
train['Year'] = train['Date'].dt.year
train['Month'] = train['Date'].dt.month
train['Week'] = train['Date'].dt.isocalendar().week.astype('int')

train['IsHoliday'] = train['IsHoliday'].apply(lambda x: 1 if x==True else 0)


promotion_cols = ['Promotion1', 'Promotion2', 'Promotion3', 'Promotion4', 'Promotion5']
for col in promotion_cols:
    train[col] = train[col].fillna(train[col].mean())

train['Total_Promotion'] = train[promotion_cols].sum(axis=1)
train = train.drop(columns=promotion_cols)

def remove_outliers_iqr(df, columns):
    z_scores = np.abs(stats.zscore(df[columns]))
    threshold = 3
    return df[(z_scores < threshold).all(axis=1)]
    
continuous_features = ['Temperature', 'Fuel_Price', 'Unemployment', 'Weekly_Sales']

train = remove_outliers_iqr(train, continuous_features)

# train['yearly_fuel_price'] = train['Year'] * train['Fuel_Price']

train = train[~train['Month'].isin([11, 12])]

train = train.drop(['id', 'Date'], axis=1)
train.head()
```

train 피처의 상관계수를 알아보자.

Weekly_Sales랑 상관계수가 높은 것은 그나마 Store, Promotion 5개의 피처, Unemployment 정도로 보인다.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12,6]
sns.heatmap(train.corr(), annot=True)
```

StandardScaler로 정규화를 진행해보자.

```python
from sklearn.preprocessing import StandardScaler

numeric_features = ['Temperature', 'Unemployment', 'Total_Promotion']
scaler = StandardScaler()
train[numeric_features] = scaler.fit_transform(train[numeric_features])
```

```python
X = train.drop(['Weekly_Sales'], axis=1)
y = train['Weekly_Sales']
```

모델 정의 후 KFold로 교차 검증을 진행해보자.

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import VotingRegressor, RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
xgb = XGBRegressor(random_state=42)
lgb = LGBMRegressor(random_state=42)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import root_mean_squared_error
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)

rmse_scores = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    best_model = VotingRegressor(estimators=[
    ('rf', rf),
    ('xgb', xgb),
    ('lgb', lgb)])
    best_model.fit(X_train, y_train)

    y_pred = best_model.predict(X_test)
    rmse = root_mean_squared_error(y_test, y_pred)

    rmse_scores.append(rmse)
```

RMSE를 측정해보자.

```python
# 각 폴드에서의 rmse 출력
print("각 폴드에서의 정확도:", rmse_scores)

# 평균 rmse 계산 및 출력
print("평균 정확도:", np.mean(rmse_scores))
```

test에도 동일하게 전처리를 진행해보자.

```python
test['Date'] = pd.to_datetime(test['Date'], format="%d/%m/%Y")
test['Year'] = test['Date'].dt.year
test['Month'] = test['Date'].dt.month
test['Week'] = test['Date'].dt.isocalendar().week.astype('int')

test['IsHoliday'] = test['IsHoliday'].apply(lambda x: 1 if x==True else 0)

test['Total_Promotion'] = test[promotion_cols].sum(axis=1)
test = test.drop(columns=promotion_cols)

# test['yearly_fuel_price'] = test['Year'] * test['Fuel_Price']

test[numeric_features] = scaler.fit_transform(test[numeric_features])

test = test.drop(['id', 'Date'], axis=1)
```

```python
train = pd.read_csv('data/shopping/train.csv')
promotion_cols = ['Promotion1', 'Promotion2', 'Promotion3', 'Promotion4', 'Promotion5']
train[promotion_cols] = train[promotion_cols].fillna(0)
train['Total_Promotion'] = train[promotion_cols].sum(axis=1)
train_zero_promo = train[train['Total_Promotion'] == 0]
train_nonzero_promo = train[train['Total_Promotion'] > 0]

features_org = ['Store', 'Temperature', 'Fuel_Price', 'Unemployment']

# 특성들의 분포를 시각화하기 위한 함수
def plot_feature_distribution(df1, df2, label1, label2, features):
    i = 0
    sns.set_style('whitegrid')
    plt.figure(figsize=(14,8))
    for feature in features:
        i += 1
        plt.subplot(3,3,i)
        sns.kdeplot(df1[feature], bw_adjust=0.5, label=label1)
        sns.kdeplot(df2[feature], bw_adjust=0.5, label=label2)
        plt.title(feature, fontsize=14)
        plt.legend()
    plt.tight_layout()
    plt.show()

# train_normal과 train_abnormal 데이터셋의 특성 분포 시각화
plot_feature_distribution(train_nonzero_promo, train_zero_promo, "train_nonzero", "train_zero", features_org)
```
